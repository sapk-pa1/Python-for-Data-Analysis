```python
#default values 
'''
It’s very common to have logic like:
if key in some_dict:
    value = some_dict[key]
else:
    value = default_value
Thus, the dict methods get and pop can take a default value to be returned, so that the above if-else block 
can be written simply as:
value = some_dict.get(key, default_value)
'''
words = ['apple', 'bat', 'bar', 'atom', 'book']
by_letter = {}
for word in words: 
    # this is taking the first letter of the word in a certain iteration 
    letter = word[0]
    if letter not in by_letter:
    #if letter is not in the by_letter dict then assign it to element of the dict with the key letter 
        by_letter[letter] = [word]
    else:
    # else add another elements to the particular key 
        by_letter[letter].append(word)
by_letter
```

```python
#The setdefault dict method is for precisely this purpose.
by_letter = {}
for word in words: 
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)
by_letter
```

# Currying: Partial Argument Application
i like my curry how i like my women , spicy! LOL. 

```python
'''
Currying is computer science jargon (named after the mathematician Haskell Curry)
that means deriving new functions from existing ones by partial argument application
'''
def add_numbers (x,y):
    return x + y 
add_5 = lambda y: add_numbers(5, y)
# the second argument to add_numbers is said to be curried .
```

```python
add_5(1)
```

```python
# the built in function partial can do the same thing 
from functools import partial 
add_five = partial(add_numbers, 5 )
add_five(12)
```

# Generators 
Generating what ? 

```python
'''
Having a consistent way to iterate over sequences, like objects in a list or lines in a
file, is an important Python feature. This is accomplished by means of the iterator
protocol, a generic way to make objects iterable. For example, iterating over a dict
yields the dict keys:
'''
some_dict = {'a': 1, 'b': 2, 'c': 3}
for key in some_dict: 
    print(key)
```

```python
'''
A generator is a concise way to construct a new iterable object. Whereas normal func‐
tions execute and return a single result at a time, generators return a sequence of
multiple results lazily, pausing after each one until the next one is requested. To create
a generator, use the yield keyword instead of return in a function:
'''
def squares(n = 10): 
    print("Generating squares from 1 to {0}".format(n**2))
    for i in range(1, n+1):
        yield i **2
'''
Yield is a keyword in Python that is used to return from a function without destroying the states of its 
local variable and when the function is called, the execution starts from the last yield statement. 
Any function that contains a yield keyword is termed a generator.
'''
gen = squares()
```

```python
gen
```

```python
print(list(gen))
```

```python
#itertools module
'''
The standard library itertools module has a collection of generators for many com‐
mon data algorithms. For example, groupby takes any sequence and a function,
grouping consecutive elements in the sequence by return value of the function. Here’s
an example:
'''
import itertools 
first_letter  = lambda x : x[0]
names = ['Alan', 'Adam', 'Wes', 'Will', 'Albert', 'Steven']
for letter , name in itertools.groupby(names, first_letter):
    print(letter , list(name))
```
