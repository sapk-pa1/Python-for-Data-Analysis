```python
#default values 
'''
It’s very common to have logic like:
if key in some_dict:
    value = some_dict[key]
else:
    value = default_value
Thus, the dict methods get and pop can take a default value to be returned, so that the above if-else block 
can be written simply as:
value = some_dict.get(key, default_value)
'''
words = ['apple', 'bat', 'bar', 'atom', 'book']
by_letter = {}
for word in words: 
    # this is taking the first letter of the word in a certain iteration 
    letter = word[0]
    if letter not in by_letter:
    #if letter is not in the by_letter dict then assign it to element of the dict with the key letter 
        by_letter[letter] = [word]
    else:
    # else add another elements to the particular key 
        by_letter[letter].append(word)
by_letter
```

```python
#The setdefault dict method is for precisely this purpose.
by_letter = {}
for word in words: 
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)
by_letter
```

```python
from collections import defaultdict
by_letter= defaultdict(list)
for word in words:
    by_letter[word[0]].append(word)
```

```python
# List , Dict and Set Comprehension 
'''
List comprehensions are one of the most-loved Python language features. They allow
you to concisely form a new list by filtering the elements of a collection, transforming
the elements passing the filter in one concise expression. They take the basic form:
[expr for val in collection if condition]
'''
strings = ['a', 'as', 'bat', 'car', 'dove', 'python']
new_string = [x.upper() for x in strings if len(x)>2]
```

```python
new_string
```

```python
'''
Set and dict comprehensions are a natural extension, producing sets and dicts in an
idiomatically similar way instead of lists. A dict comprehension looks like this:
dict_comp = {key-expr : value-expr for value in collection if condition}
'''
loc_mapping = {val : index for index , val in enumerate(strings)}
```

```python
states = ['Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda','southcarolina##', 'West virginia?']
import re 
def clean_string(string):
    result = []
    for value in string: 
        value = value.strip()
        value = re.sub('[#?&!]', '', value)
        value = value.title()
        result.append(value)
    return result
```

```python
cleaned_state = clean_string(states)
cleaned_state
```

# Anonymous Lambda Function 

```python
def square(x):
    return x*x
print(square(10))
```

```python
#similar implementation of it in lambda 
result = lambda x:x*x  
#  parameter  ->| |-> Return value 
print(result(10) )
```

One reason lambda functions are called anonymous functions is
that , unlike functions declared with the def keyword, the function
object itself is never given an explicit '__name__' attribute.
